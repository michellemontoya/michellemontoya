```python
class Persona():
  def init(self):
    self.__Nombre=""
    self.__Apellido=""
    self.__Edad=0

  def GetNombre(self):
    return self.__Nombre

  def SetNombre(self, nombre):
    self.__Nombre=nombre
  
  def GetApellido(self):
    return self.__Apellido

  def SetApellido(self, apellido):
    self.__Apellido = apellido

  def GetEdad(self):
    return self.__Edad

  def SetEdad(self, edad):
    self.__Edad=edad


#Alumno
  
class Alumno(Persona):
  def init(self):
    self.__Curso=""

    self.__Asignatura=""

  def GetCurso(self):
    return self.__Curso

  def SetCurso(self, curso):
    self.__Curso= curso

  def GetAsignatura(self):
    return self.__Asignatura
  
  def SetAsignatura(self, asignatura):
    self.__Asignatura = asignatura 

  def MostrarAlumno(self):
    print("Alumno: ")
    print("\tNombre:",self.GetNombre())
    print("\tApellido:",self.GetApellido())
    print("\tEdad:",self.GetEdad())
    print("\tCurso:",self.__Curso)
    print("\tMatricula:",self.__Asignatura)
    

alumno=Alumno()
alumno.SetNombre("jorge ")
alumno.SetApellido("henao")
alumno.SetEdad("30")
alumno.SetCurso("bachillerato")
alumno.SetAsignatura(["matematicas,"  "tecnologia"])
alumno.MostrarAlumno()

print("///////////////////////////////////////////////////////////////////////////")

#Profesor
class Profesor(Persona):
  def init(self):
    self.__Antiguedad=""
    self.__Tutorias=""
    self.__Telefono=""

  def GetAntiguedad(self):
    return self.__Antiguedad

  def SetAntiguedad(self, antiguedad):
    self.__Antiguedad = antiguedad

  def GetTutorias(self):
    return self.__Tutorias

  def SetTutorias(self, tutorias):
    self.__Tutorias = tutorias

  def GetTelefono(self):
    return self.__Telefono

  def SetTelefono(self, telefono):
    self.__Telefono = telefono

  def MostrarProfesor(self):
    print("Profesor: ")
    print("\tNombre:",self.GetNombre())
    print("\tApellido:",self.GetApellido())
    print("\tEdad:",self.GetEdad())
    print("\tAntiguedad:",self.__Antiguedad)
    print("\tTutorias:",self.__Tutorias)
    print("\tTelefono:",self.__Telefono)


profesor = Profesor()
profesor.SetNombre("jorge")
profesor.SetApellido("henao")
profesor.SetEdad(45)
profesor.SetAntiguedad(30)
profesor.SetTutorias([[ "lunes:"+"2-4 pm","martes:"+"4-6 pm"]])
profesor.SetTelefono("3124500000")
profesor.MostrarProfesor()

print("///////////////////////////////////////////////////////////////////////////")

#Investigador:
class Investigador(Persona):
  def init(self):
    
    self.__Especialidad=""
    self.__Años=""

  

  def GetEspecialidad(self):
    return self.__Especialidad

  def SetEspecialidad(self, especialidad):
    self.__Especialidad = especialidad

  def GetAños(self):
    return self.__Años

  def SetAños(self, años):
    self.__Años = años

  def MostrarInvestigador(self):
    print("Investigador: ")
    print("\tNombre:",self.GetNombre())
    print("\tApellido:",self.GetApellido())
    print("\tEdad:",self.GetEdad())
    print("\tEspecialidad:",self.__Especialidad)
    print("\tAños:",self.__Años)


investigador = Investigador()
investigador.SetNombre("jorge ")
investigador.SetApellido("henao")
investigador.SetEdad(45)
investigador.SetEspecialidad("detective")
investigador.SetAños(4)
investigador.MostrarInvestigador()

print("///////////////////////////////////////////////////////////////////////////")
#Profesor_Universidad

class Profesor_Universidad(Persona):
  def init(self):
    self.__Universidad=""
    self.__Departamento=""

  def GetUniversidad(self):
    return self.__Universidad

  def SetUniversidad(self, universidad):
    self.__Universidad = universidad

  def GetDepartamento(self):
    return self.__Departamento

  def SetDepartamento(self, departamento):
    self.__Departamento = departamento

  def MostrarProfesor_Universidad(self):
    print("Profesor de Universidad: ")
    print("\tNombre:",self.GetNombre())
    print("\tApellido:",self.GetApellido())
    print("\tEdad:",self.GetEdad())
    print("\tUniversidad:",self.__Universidad)
    print("\tDepartamento:",self.__Departamento)


profesor_universidad = Profesor_Universidad()
profesor_universidad.SetNombre("jorge")
profesor_universidad.SetApellido("henao")
profesor_universidad.SetEdad(30)
profesor_universidad.SetUniversidad("unidades tecnologicas de santander ")
profesor_universidad.SetDepartamento("santander-bucaramanga")
profesor_universidad.MostrarProfesor_Universidad()
```

```python
#punto 3

class Animal():
  def init(self):
    self.__Color=""
    self.__Peso=""
    self.__Comer=""
    self.__Moverse=""
   

  def GetColor(self):
    return self.__Color

  def SetColor(self, color):
    self.__Color= color
  
  def GetPeso(self):
    return self.__Peso

  def SetPeso(self, peso):
    self.__Peso = peso

  def GetComer(self):
    return self.__Comer

  def SetComer(self, comer):
    self.__Comer = comer

  def GetMoverse(self):
    return self.__Moverse

  def SetMoverse(self, moverse):
    self.__Moverse = moverse


class Ave(Animal):
  def init(self):
    self.__Tipo=""
    self.__Volar=""
    

  def GetTipo(self):
    return self.__Tipo

  def SetTipo(self, tipo):
    self.__Tipo=tipo
  
  def GetVolar(self):
    return self.__Volar

  def SetVolar(self, volar):
    self.__Volar = volar

  def MostrarAve(self):
    print("ave: ")
    print("\tColor:",self.GetColor())
    print("\tPeso:",self.GetPeso())
    print("\tComer:",self.GetComer())
    print("\tMoverse:",self.GetMoverse())
    print("\tTipo:",self.GetTipo())
    print("\tVolar:",self.GetVolar())

ave=Ave()
ave.SetColor("aguamarina")
ave.SetPeso("2.7 kg")
ave.SetComer("gusanos")
ave.SetMoverse("aleteo")
ave.SetTipo("vertebrado")
ave.SetVolar("veloz")
ave.MostrarAve()

class Perro(Animal):
  def init(self):
    self.__Raza=""
    self.__Ladra=""
    

  def GetRaza(self):
    return self.__Raza

  def SetRaza(self, raza):
    self.__Raza=raza
  
  def GetLadra(self):
    return self.__Ladra

  def SetLadra(self, ladra):
    self.__Ladra = ladra

  def MostrarPerro(self):
    print("perro: ")
    print("\tColor:",self.GetColor())
    print("\tPeso:",self.GetPeso())
    print("\tComer:",self.GetComer())
    print("\tMoverse:",self.GetMoverse())
    print("\tRaza:",self.GetRaza())
    print("\tLadra:",self.GetLadra())

perro=Perro()
perro.SetColor("blanco con cafe ")
perro.SetPeso("12 kg")
perro.SetComer("cocentrado")
perro.SetMoverse("en 4 patas")
perro.SetRaza("chitzu")
perro.SetLadra("fuerte")
perro.MostrarPerro()    


  
```
